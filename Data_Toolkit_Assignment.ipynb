```python
# 2.write a pandas script to find the mean of a specific columm in a dataframe.
import pandas as pd
df=pd.DataFrame({
    'name':['ankita','punit','ridhi'],
     'age':[20,21,22],
    'gender':['female','male','female']
})
print(df)
print(df['age'].mean())
```

```python
# 4. how do you calculate the correlation matrix using seaborn and visualize i twith a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
# select only numerical features for correlation calculation
numerical_df=df.select_dtypes(include=['number'])
corr=numerical_df.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# 5. generate a bar plot using plotly.
import plotly.express as px
df=px.data.tips()
fig=px.bar(df,x='day',y='total_bill',color='sex')
fig.show()
```

```python

# 6. create a dataframe and add a new columm based on an exisiting columm.
import pandas as pd
df=pd.DataFrame({
    'name':['ankita','punit','ridhi'],
    'age':[20,21,22],
    'gender':['female','male','female']
})
df['age_group']=df['age'].apply(lambda x: 'young' if x<25 else 'old')
print(df)
```

```python
# 9. generate a pandas dataframe and filterous where a columm value is greater than a theshold.
import pandas as pd
df=pd.DataFrame({
    'name':['ankita','punit','ridhi'],
    'age':[20,21,22],
    'gender':['female','male','female']
})
df=df[df['age']>20]
print(df)

```
